```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
voice_data = pd.read_csv('voice.csv')
```

```python
voice_data.head()
```

```python
voice_data.isnull()
```

```python
voice_data.shape
```

```python
voice_data['label'].value_counts()
```

```python
X = voice_data.drop (columns='label',axis=1)
Y = voice_data['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accurasy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('akurasi data training :', training_data_accurasy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accurasy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('akurasi data testing :', test_data_accurasy)
```
